```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Data Cleaning of IPL_Ball_By_Ball Table

```python
IPL_Ball_By_Ball = pd.read_csv( r'C:\Users\admin\Desktop\Practice Assignment for Python\IPL Ball-by-Ball 2008-2020.csv')
IPL_Ball_By_Ball.head(5)
```

```python
df = pd.DataFrame(IPL_Ball_By_Ball)
print(f"The Total Rows & Column has", df.shape)
df.info()
```

```python
duplicate_data = df[df.duplicated()]
print("The Duplicate Data count is:", len(duplicate_data))
df = df.drop_duplicates()

# checking null values in column 
missing_values = df.isnull().sum()
print(f"The missing values are as below :\n{missing_values}")

# Removing missing less values from bowling_team
df = df.dropna(subset=['bowling_team'])

# filling missing values with None
df['dismissal_kind']=df['dismissal_kind'].fillna('None')
df['player_dismissed']=df['player_dismissed'].fillna('None')
df['fielder']=df['fielder'].fillna('None')
df['extras_type']=df['extras_type'].fillna('None')
```

# Data Cleaning of IPL_Matches Table

```python
IPL_Matches = pd.read_csv(r'C:\Users\admin\Desktop\Practice Assignment for Python\IPL Matches 2008-2020.csv')
IPL_Matches.head(5)
```

```python
df_1 = pd.DataFrame(IPL_Matches)

print(f"The Total Rows & Column has ", df_1.shape)

df_1.info()
```

```python
# changing datatype of date column
df_1['date']=pd.to_datetime(df_1['date'])

# changing datatype of result_marging column
# df_1['result_margin']=df_1['result_margin'].astype(int)

# checking again datatype of date column
df_1.dtypes
```

```python
# checking missing values count
missing_values = df_1.isnull().sum()
print(f"The Missing Values are as below:\n{missing_values}")

# filling missing_values of City Column
df_1['city'].fillna(df_1['city'].mode()[0],inplace=True)

# filling missing_values of player_of_match column
df_1['player_of_match']=df_1['player_of_match'].fillna('None')

# filling missing values of result column
df_1['result']=df_1['result'].fillna('None')

# filling missing values of winner column
df_1['winner']=df_1['winner'].fillna('None')

# filling missing values of eliminator column
df_1['eliminator']=df_1['eliminator'].fillna('None')

# filling missing values of result margin column
df_1['result_margin']=df_1['result_margin'].fillna('None')

# filling missing values of method column
df_1['method']=df_1['method'].fillna('None')
```

# Basic Match IPL Analysis

```python
# Total Matches
total_matches = df_1['date'].value_counts().sum()
print(f"The Total Matches is: ", total_matches)

# Total Teams
total_teams = pd.concat([df_1['team1'],df_1['team2']]).nunique()
print(f"The Total Team is: ", total_teams)

# Total Venues
total_venues = df_1['venue'].nunique()
print(f"The Total Venue is: ",total_venues)

# Total Cities
total_cities = df_1['city'].nunique()
print(f"The Total City is: ", total_cities)
```

# Toss Decisions Across Seasons

```python
# Extracting year from date
df_1['season']=pd.to_datetime(df_1['date']).dt.year

# Group By season and toss_decision
toss_decision_seson = df_1.groupby(['season','toss_decision'])['id'].count().unstack().fillna(0)

toss_decision_seson.plot(kind ='bar',stacked = True, figsize= (12,6), colormap = 'Set2')
plt.title('Toss Decisions Across Season')
plt.xlabel('Season')
plt.ylabel('count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Does Winning Toss Mean Winning the Game?

```python
# Matches where toss winner = match winner
toss_win_match_win = df_1[df_1['toss_winner']==df_1['winner']].shape[0]

# Total Matches with a winner
valid_matches = df_1[df_1['winner'].notna()].shape[0]

# calculate %
ratio = (toss_win_match_win /valid_matches)*100
print(f"Toss winner also won the match in {ratio:.2f}% of games.")
```

# Which Teams Have Won the Tournament the Most

```python
# Grouping by winner and getting the count of wins
winning_team_count = df_1['winner'].value_counts()

# Display winning teams and their win counts
print("Winning Team and Their Count of Wins:")
print(winning_team_count)

# Plotting the winning teams' count
plt.bar(winning_team_count.index, winning_team_count.values)
plt.xlabel('Winning Team')
plt.ylabel('Won Matches Count')
plt.title('Teams and Their Number of Wins')

# Adding labels on top of the bars
for i, value in enumerate(winning_team_count.values):
    plt.text(i, value + 1, str(value), ha='center', va='bottom')

# Rotate x-axis labels for better readability
plt.xticks(rotation=90)

# Layout adjustment
plt.tight_layout()
plt.show()
```

# Comparative Analysis of Teams

```python
# Total matches played by each team
all_teams = pd.concat([df_1['team1'], df_1['team2']])
matches_played = all_teams.value_counts()

# Total matches won
matches_won = df_1['winner'].value_counts()

# Combine in a single DataFrame
team_comparison = pd.DataFrame({
    'Matches Played': matches_played,
    'Matches Won': matches_won
})
team_comparison['Win %'] = (team_comparison['Matches Won'] / team_comparison['Matches Played']) * 100

print(team_comparison.sort_values('Win %', ascending=False).round(2))
```
